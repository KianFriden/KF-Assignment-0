```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
sin=[]
cos=[]
for x in range(360):
    sin+=[np.sin(x*np.pi/180)]
    cos+=[np.cos(x*np.pi/180)]
    
plt.plot(sin)
plt.plot(cos)
plt.title('Cosine and Sine')
plt.ylabel('value')
plt.xlabel('Degrees')
```
